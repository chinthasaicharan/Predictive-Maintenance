```python
import sklearn as sk
from sklearn.model_selection import train_test_split
```

```python
from sklearn import svm
```

```python
import numpy as np
```

```python
X_train,X_test,y_train,y_test = train_test_split(data.iloc[:,:-1],
                                                data.iloc[:,-1]
                                                ,test_size=0.4,random_state=8)
```

```python
clf = svm.OneClassSVM(nu=0.1, kernel="rbf", gamma=0.5)

# Fit the SVM model to the data
clf.fit(X_train)
```

```python
# Predict the anomalies
y_pred = clf.predict(X_test)

# Print the number of anomalies
print("Number of anomalies: ", len(np.where(y_pred == -1)[0]))
```

```python
from sklearn.cluster import DBSCAN



# Define the DBSCAN model
model_4 = DBSCAN(eps=100, min_samples=5, metric='euclidean')

# Fit the model to the data
model_4.fit(X_train)

# Predict the anomalies
y_pred_4 = np.array([1 if x == -1 else 0 for x in model_4.labels_])

# Print the number of anomalies
print("Number of anomalies: ", len(np.where(y_pred_4 == 1)[0]))
```

<h1> Using NAIVE BAYES </h1>

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix

model_5 = GaussianNB()

# Fit the model to the data
model_5.fit(X_train, y_train)

# Predict the anomalies
y_pred_5 = model_5.predict(X_test)

# Print the number of anomalies
print("Number of anomalies: ", len(np.where(y_pred_5 == 1)[0]))
```

```python
print(f"f1 score : {f1_score(y_test,y_pred_5)}")
```

```python
print(f"accuracy : {accuracy_score(y_test,y_pred_5)}")
```

```python
print(f"confusion_matrix : {confusion_matrix(y_test,y_pred_5)}")
```
